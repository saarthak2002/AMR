```python
#Data import and set up

import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None

YEAR1 = 1999
YEAR2 = 2000

StaphAureuaDataMRSA1 = pd.read_csv('dividedData/StaphAureuaMRSA{}.csv'.format(YEAR1))
StaphAureuaDataMSSA1 = pd.read_csv('dividedData/StaphAureuaMSSA{}.csv'.format(YEAR1))

StaphAureuaDataMRSA1.drop(['Organism', 'PatientAge', 'PatientGender', 'PatientLocation', 'Region', 'State', 'Zipcode',
                      'SpecimenSource', 'SpecimenSourceCategory','ReportDate'], axis=1, inplace=True)
StaphAureuaDataMSSA1.drop(['Organism', 'PatientAge', 'PatientGender', 'PatientLocation', 'Region', 'State', 'Zipcode',
                      'SpecimenSource', 'SpecimenSourceCategory','ReportDate'], axis=1, inplace=True)

StaphAureuaDataMRSA2 = pd.read_csv('dividedData/StaphAureuaMRSA{}.csv'.format(YEAR2))
StaphAureuaDataMSSA2 = pd.read_csv('dividedData/StaphAureuaMSSA{}.csv'.format(YEAR2))

StaphAureuaDataMRSA2.drop(['Organism', 'PatientAge', 'PatientGender', 'PatientLocation', 'Region', 'State', 'Zipcode',
                      'SpecimenSource', 'SpecimenSourceCategory','ReportDate'], axis=1, inplace=True)
StaphAureuaDataMSSA2.drop(['Organism', 'PatientAge', 'PatientGender', 'PatientLocation', 'Region', 'State', 'Zipcode',
                      'SpecimenSource', 'SpecimenSourceCategory','ReportDate'], axis=1, inplace=True)
```

```python
StaphAureuaDataMRSA1 = StaphAureuaDataMRSA1.add_suffix('_{}'.format(YEAR1))
StaphAureuaDataMSSA1 = StaphAureuaDataMSSA1.add_suffix('_{}'.format(YEAR1))

StaphAureuaDataMRSA2 = StaphAureuaDataMRSA2.add_suffix('_{}'.format(YEAR2))
StaphAureuaDataMSSA2 = StaphAureuaDataMSSA2.add_suffix('_{}'.format(YEAR2))
```
