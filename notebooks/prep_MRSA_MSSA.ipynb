```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
def prepareDataMRSA(region):
    
    for YEAR in range(1999,2013):
        StaphAureuaDataPath = 'splitData/{}{}.csv'.format(region.replace(' ',''),YEAR)
    
        try:
            StaphAureuaData = pd.read_csv(StaphAureuaDataPath)
        except OSError as e:
            continue
            
        StaphAureuaDataMRSA = StaphAureuaData[ (StaphAureuaData['Oxacillin'] == 'R') | (StaphAureuaData['Oxacillin'] == 'I') ]
        StaphAureuaDataMRSA = StaphAureuaDataMRSA.reset_index(drop=True)
        StaphAureuaDataMRSA.drop("Oxacillin", axis=1, inplace=True)
        StaphAureuaDataMRSA.drop(['Organism', 'PatientAge', 'PatientGender', 'PatientLocation', 'Region', 'State', 'Zipcode',
                                  'SpecimenSource', 'SpecimenSourceCategory','ReportDate','HSA_REGION'], axis=1, inplace=True)
        associationDict = {
            'Amoxicillin/clavulanate' : 1,
            'Ampicillin' : 2,
            'Ampicillin/sulbactam' : 3,
            'Cefazolin' : 4, 
            'Ceftriaxone' : 5,
            'Chloramphenicol' : 6,
            'Ciprofloxacin' : 7,
            'Clindamycin' : 8,
            'Erythromycin' : 9,
            'Gentamicin' : 10,
            'Imipenem' : 11,
            'Levofloxacin' : 12,
            'Linezolid' : 13,
            'Moxifloxacin' : 14,
            'Nitrofurantoin' : 15,
            'Penicillin' : 16,
            'Quinupristin/dalfopristin' : 17,
            'Rifampin (Rifampicin)' : 18,
            'Tetracycline' : 19,
            'Trimeth/sulfa' : 20,
            'Vancomycin' : 21
        }
        i = 0
        for column in StaphAureuaDataMRSA.columns[0:21]:
            for item in StaphAureuaDataMRSA[column]:
                if(item == "R" or item == "I"):
                    StaphAureuaDataMRSA[column][i] = associationDict[column]
                else:
                    StaphAureuaDataMRSA[column][i] = ' '
                i=i+1
            i=0
        
        inputFileName = 'tempMRSA/{}{}MRSA.txt'.format(region.replace(' ',''),YEAR)
        finalName = '{}_{}_MRSA.txt'.format(region.replace(' ',''),YEAR)
        outputFileName = '/home/uzn2up/MRSA_MSSA_output/{}_MRSA/{}'.format(region.replace(' ',''),finalName)

        with open(inputFileName, 'a') as f:
            dfAsString = StaphAureuaDataMRSA.to_string(header=False, index=False, na_rep='')
            f.write(dfAsString)

        with open(inputFileName,'r+') as file:
            for idx, line in enumerate(file):
                if not line.isspace():
                    with open(outputFileName,'a') as file2:
                        file2.write(line)
        with open(outputFileName,'a') as file3:
            file3.write('\n')
        with open(outputFileName, 'r') as f2:
            tmp = list(f2)
            if tmp[-1] == '\n':
                tmp.pop()
                with open(outputFileName, 'w+') as fd:
                    fd.seek(0)                  
                    fd.truncate()
                    for line in tmp:
                        fd.write(line)
```
