```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
distMat = pd.read_csv('distMat.csv',index_col=0)
```

```python
distMat
```

```python
plt.figure(figsize=(120,120))
ax = sns.heatmap(distMat,cmap="Spectral_r",linewidths=1,linecolor='white')

cbar = ax.collections[0].colorbar
# here set the labelsize by 20
cbar.ax.tick_params(labelsize=50)
fig = ax.figure
fig.savefig('distmat.pdf',dpi=300)
```
