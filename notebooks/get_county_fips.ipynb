```python
import pandas as pd
allData = pd.read_csv('/project/biocomplexity/anil/AMR/StaphAureusAntibiogramTimeSpace/RawData/StaphAureua2012.csv',converters={'Zipcode': lambda x: str(x)})
allData['Zipcode'] = allData['Zipcode'].str.zfill(5)
```

```python
i = 0
for code in allData['Zipcode']:
    j = 0
    for code2 in zipstofips['ZIP']:
        
        if(code2 == code):
            allData['fips'][i] = zipstofips['STCOUNTYFP'][j]
        
        j = j + 1
        
    
    i = i + 1
```
