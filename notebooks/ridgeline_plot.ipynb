```python
#Data import and set up

import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None

YEAR = 2012

StaphAureuaDataPath = '/project/biocomplexity/anil/AMR/StaphAureusAntibiogramTimeSpace/RawData/StaphAureua{}.csv'.format(YEAR)
hsaRegionDataPath = '/home/uzn2up/crosswalks/ZipHsaHrr{}.xls'.format(YEAR)

StaphAureuaData = pd.read_csv(StaphAureuaDataPath)
hsaRegionData = pd.read_excel(hsaRegionDataPath)
```

```python
associationDictYear = {
    1999 : '22',
    2000 : '23',
    2001 : '24',
    2002 : '25',
    2003 : '26',
    2004 : '27',
    2005 : '28',
    2006 : '29',
    2007 : '30',
    2008 : '31',
    2009 : '32',
    2010 : '33',
    2011 : '34',
    2012 : '35'
}
```

```python
StaphAureuaDataMRSA = StaphAureuaData[ (StaphAureuaData['Oxacillin'] == 'R') | (StaphAureuaData['Oxacillin'] == 'I') ]
StaphAureuaDataMSSA = StaphAureuaData[ (StaphAureuaData['Oxacillin'] == 'S') | (pd.isnull(StaphAureuaData['Oxacillin'])) ]
StaphAureuaDataMRSA.drop("Oxacillin", axis=1, inplace=True)
StaphAureuaDataMSSA.drop("Oxacillin", axis=1, inplace=True)
```

```python
StaphAureuaDataMRSA.to_csv('dividedData/StaphAureuaMRSA{}.csv'.format(YEAR), index=False)
StaphAureuaDataMSSA.to_csv('dividedData/StaphAureuaMSSA{}.csv'.format(YEAR), index=False)
```
